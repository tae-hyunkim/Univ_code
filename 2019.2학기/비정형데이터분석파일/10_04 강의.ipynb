```python
import pandas as pd

text = []
score = []
name = []

for i in range(1,6):
    url = 'https://movie.daum.net/moviedb/grade?movieId=133315&type=netizen&page={}'.format(i)
    req = requests.get(url).text
    root = lxml.html.fromstring(req)
    for j in root.cssselect('em.link_profile'):
        name.append(j.text_content())
    for j in root.cssselect('p.desc_review'):
        text.append(j.text_content().strip().replace('\r',''))
    for j in root.cssselect('em.emph_grade'):
        score.append(j.text_content())
pd.DataFrame({'name':name,'score':score,'review':text})
```

```python
text = []
score = []
name = []
for i in range(1,6):
    url = 'https://movie.daum.net/moviedb/grade?movieId=133315&type=netizen&page=' + str(i)
    req = requests.get(url).text
    root = lxml.html.fromstring(req)
    named = root.cssselect('em.link_profile')
    texted = root.cssselect('p.desc_review')
    scored = root.cssselect('em.emph_grade')
    for j,k,l in zip(named,texted,scored):
        name.append(j.text_content())
        text.append(k.text_content().strip().replace('\r',''))
        score.append(l.text_content())

pd.DataFrame({'reviewer':name,'score':score,'review':text})
```

```python
# 결과물을 보면 현재 상대주소 형태로 나와있음을 알 수 있음 이는 해당 url만을 사용하였을 때 사이트에 접속이 되지 않는다는 의미임. 
```

```python
# 상대주소 절대주소로 바꾸기
import pandas as pd
import urllib.parse # url의 값을 입력하면 해당 url을 베이스로 한 절대주소를 만들어주는 패키지임. 
url = 'https://m.cafe.naver.com/ArticleList.nhn?search.clubid=19773565&search.menuid=15&search.boardtype=L'
req = requests.get(url)
root = lxml.html.fromstring(req.text)

urls = []
title = []
for i,j in zip(root.cssselect('a.txt_area'),root.cssselect('strong.tit')):
    urls.append(urllib.parse.urljoin(url,i.attrib['href'])) # Base URL에 나온 상대주소값 삽입을 통해 URL 추출 
    title.append(j.text_content().strip())
df = pd.DataFrame({'urls':urls,'title':title})
df.head()
```

```python
lxml.html.fromstring(requests.get(urls[0]).text).cssselect('#postContent')
# 보다시피 requests.get()을 그냥 활용해서 가져오면 접근이 되지 않는다는 사실을 알 수 있음
# 특정 검색어로 검색을 한 후 거기에서 나오는 결과물을 클릭하면 네이버 카페에 가입하지 않아도 접속할 수 있다는 사실을 활용하여
# 접근하기로 함 이때 사용하는 함수는 headers={'referer':검색URL}이다. 
```

```python
# 제대로 돌아가는 것을 확인 
lxml.html.fromstring(requests.get(urls[0],
                                  headers = {'referer':
                                             'https://search.naver.com/search.naver?sm=top_hty&fbm=0&ie=utf8&query=%EC%89%B4%EB%95%8C+%ED%8E%B8%ED%95%9C+%EC%9D%98%EC%9E%90+%EC%B6%94%EC%B2%9C+%EB%B6%80%ED%83%81%EB%93%9C%EB%A0%A4%EC%9A%94.'}
                                 ).text).cssselect('#postContent')[0].text_content().strip() # 리스트로 나오기 때문에 [0]취해줌 
```

```python
body = []
out = []
for i in range(len(urls)):
    req = requests.get(urls[i],headers={'referer':'https://search.naver.com/search.naver?sm=top_hty&fbm=0&ie=utf8&query=%EC%95%84%EB%AC%B4%EA%B1%B0%EB%82%98'
                                   })
    # referer 어떤 것으로 들어가서 접속했는지에 대한 정보 제공하는 방법
    # naver의 검색창에서 접속했다고 하면 됨. 
    root = lxml.html.fromstring(req.text)
    if '카페 멤버만 볼 수 있습니다.' in root.text_content() :
        print(i)
        out.append(i)
    for j in root.cssselect('#postContent'):
        body.append(j.text_content().strip())
for i in out:
    title.pop(i)
pd.DataFrame({'title':title,'body':body})
```
