```python
a = requests.get('https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5')
a.text
```

```python
import lxml.html

root = lxml.html.fromstring(a.text)
for i in root.cssselect('a.f_nb'):
    print(i.text_content())
    print(i.attrib['href'])
```

```python
for i in root.cssselect('.f_nb a'):
    print(i.text_content())
    print(i.attrib['href'])
```

```python
url = 'https://search.daum.net/search?w=news&DA=PGD&enc=utf8&cluster=y&cluster_page=1&q=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&p={}'
for i in range(1,5):
    res = requests.get(url.format(i))
    root = lxml.html.fromstring(res.text)
    for j in root.cssselect('.f_nb a'):
        print(j.text_content())
        print(j.attrib['href'])
```

```python
href = []
url = 'https://search.daum.net/search?w=news&DA=PGD&enc=utf8&cluster=y&cluster_page=1&q=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&p={}'
for i in range(1,5):
    res = requests.get(url.format(i))
    root = lxml.html.fromstring(res.text)
    for j in root.cssselect('.f_nb a'):
        href.append(j.attrib['href'])
href
```

```python
articles = []
for i in href:
    res = requests.get(i).text
    root = lxml.html.fromstring(res)
    for j in root.cssselect('.article_view'):
        articles.append(j.text_content())
articles[0]
```

```python
import pandas as pd

data = pd.DataFrame({'link':href, 'acrticle':articles})
data.to_csv('news.csv',index=False)
```
