```python
def extract_noun(text):
    doc = nlp(text)
    for size in range(len(doc.sentences)):
        for word in doc.sentences[size].words:
            for i,k in list(zip(word.lemma.split('+'), word.xpos.split('+'))):
                if k.startswith('N'):
                    yield i
```

```python
# yield
def series(n):
    xs = []
    for x in range(n):
        xs.append(x+1)
    return xs
series(5)
```

```python
def series(n):
    for x in range(n):
        yield x+1
# yield를 사용함으로써 빈리스트를 만들고 append 하고 return하는 것을 제외시킬수 있음. 
list(series(5))
```
