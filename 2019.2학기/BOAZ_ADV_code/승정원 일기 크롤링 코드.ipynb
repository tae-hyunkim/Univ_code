```python
import requests #데이터를 받아오는 역할
from bs4 import BeautifulSoup #데이터에서 값을 뽑아내는 역할
import pandas as pd
import time
import numpy as np
import lxml.html
import re 
import urllib.parse # url의 값을 입력하면 해당 url을 베이스로 한 절대주소를 만들어주는 패키지임
```

# 승정원 일기 크롤링 코드 작성하기 - 왕들 정보 접근

```python
url = 'http://sjw.history.go.kr/m/kinglist.do'
```

```python
req = requests.get(url, headers={
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}).text
soup = BeautifulSoup(req, 'html.parser')
```

```python
king_url = []
base_url = soup.find_all('div',{'class':'king-list'})[0].find_all('a')
for i in base_url :
    king_url.append(urllib.parse.urljoin(url,i['href']))
king_url
```

# 인조 월별 접근

```python
req = requests.get(king_url[0], headers={
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}).text
soup = BeautifulSoup(req, 'html.parser')
```

```python
month_url = []
base_url = soup.find_all('div',{'class':'month-list'})[0].find_all('a')
for i in base_url:
    month_url.append(urllib.parse.urljoin(url,i['href']))
month_url
```

# 인조 일자 접근

```python
url = month_url[0]
req = requests.get(month_url[0], headers={
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}).text
soup = BeautifulSoup(req, 'html.parser')
```

```python
day = []
base_url = soup.find_all('div',{'class':'month-list'})[1].find_all('a')
for i in base_url:
    day.append(urllib.parse.urljoin(url,i['href']))
day
```

# 인조 일자 접근 및 본문 크롤링

```python
url = day[0]
req = requests.get(day[0], headers={
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}).text
soup = BeautifulSoup(req, 'html.parser')
```

```python
original = []
for url in day:
    req = requests.get(url, headers={
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}).text
    soup = BeautifulSoup(req, 'html.parser')
    base_url = soup.find_all('div',{'class':'month-list'})[2].find_all('ul')[0].find_all('li')
    for urls in base_url:
        original.append(urls.find_all('span')[0].text.replace('\t\t\xa0\t',''))
original
```
