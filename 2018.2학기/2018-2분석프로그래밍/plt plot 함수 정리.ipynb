가장 간단한 플롯은 선을 그리는 라인 플롯(line plot)이다. 라인 플롯은 데이터가 시간, 순서 등에 따라 어떻게 변화하는지 보여주기 위해 사용한다

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r") # 인자에 대한 설명은 밑에 달아놓았음.
plt.xlim(0, 50) # x축의 범위를 0~50까지
plt.ylim(-10, 30) # y축의 범위를 -10 ~ 30 까지

# 위의 두 문장을 plt.axis((0, 50, -10, 30))로 대체해도 동일한 결과를 얻는다.

plt.show()
```

```python
# legend 이용 plot마다 이름지어진 label의 값에 따라 범례 지정 가능. loc인자를 사용 
# 범례 위치 지정해 줄 수 있음 0이 가장 좋은 위치라 생각되는 위치에 넣는값
# loc의 범위는 0~10
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
```

```python
# subplot이용 subplot의 인자는 첫번째인자 행 두번째인자 열 세번째인자 위치 지정
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(211) # (2,1,1) 과 결과 동일 
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2) # 2행 1열의 매트릭스에서 2번째 위치에 넣어라.
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

```python
# subplot이용하는 다른 방법.
fig, axes = plt.subplots(2, 2)
# fig 전체 그림 크기, axes 행렬로 나타난다. 
np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, ax0 = plt.subplots() # 그림 한개만 그릴거에요.
ax1 = ax0.twinx() # x는 겹치고 y는 축을 다르게 적용. 엑셀에서 축이 다르게 그리는것과 같다.
# twinx는 x축을 공유한다는 의미.
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
#ax0.set_ylim(0, 12)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax0.set_xlabel("공유되는 x축")
plt.show()
```

```python
# x축 y축 title에 적용되는 폰트 각각 다르게 적용하기. 
font1 = {'family': 'Malgun Gothic', 'color':  'black', 'size': 24}
font2 = {'family': 'NanumBarunpen',
         'color':  'darkred', 'weight': 'bold', 'size': 18}
font3 = {'family': 'Malgun Gothic',
         'color':  'blue', 'weight': 'light', 'size': 12}

x = np.linspace(0.0, 5.0, 100)
y = np.cos(2 * np.pi * x) * np.exp(-x)

plt.plot(x, y, 'k')
plt.title('한글 제목', font1)
plt.xlabel('엑스 축', font2)
plt.ylabel('와이 축', **font3) # dictionary 앞에 ** <- 파라미터로 나누어라. 즉 파라미터로 넘길때 양식이 바뀐다. 
# d = {'a':3} / m(**d) => m(a=3) 
# 리스트 -> 낱개로 나뉨 이때는 * 한개사용 ex) d = [1,2,3] m(*d) -> m(1,2,3)
# 값을 적용할때 딕셔너리 리스트 등의 타입이 바뀌게 된다. 
plt.subplots_adjust()
plt.show()
```

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
# 누적 bar 차트 남자와 여자 차트를 예로 들면 남자 차트 그 위에 여자 차트 그려서 해결
# button method 사용.
```

```python
# bar chart의 범례 y축으로 보냄. -> barh
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

카테고리 별 값의 상대적인 비교를 해야 할 때는 pie 명령으로 파이 차트(pie chart)를 그릴 수 있다. 파이 차트를 그릴 때는 윈의 형태를 유지할 수 있도록 다음 명령을 실행해야 한다.

plt.axis('equal')

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0) # 파이차트에서 일정 수준 띄우는 것 의미.
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%.1f%%', shadow=True, startangle=80,textprops={'fontsize': 25}) 
# autopct %% 안의 값 정규표현식 이용 .1f 소숫점 첫째자리까지 나타내라.
# textprops 이용 text크기 조절 가능. 
plt.axis('equal') # 원이 동글동글하게 나오게 만듦. 
plt.show()
```

```python
df = pd.crosstab(md['time_id']%100, md['gender'], values=md['pageviews'], aggfunc=np.size)
df.plot.bar(stacked=True,rot=0) # 쌓을 수 있음. 
# rot 이용하면 x축의 ticklabel 세워쓰기 할 수 있음. 
# plt.xticks(rotation='horizontal') 을 이용해도 세워진다. 
plt.show()
```

```python
df = pd.crosstab(md['job'], md['gender'], values=md['dwell_time'], aggfunc=np.mean)
df.index.name = None
# df의 index name 제거하면 된다. 즉 플랏에서 삭제하는 것이 아닌 데이터 프레임 상에서 제거 
df.plot.barh(subplots=True, layout=(1, 2), legend=False, sharey=True)
# Sharey 축 같이 사용할것인가? 판단. 
plt.show()
```
