```python
import pandas as pd
import numpy as np
from collections import Counter
import re
```

```python
txt_file = open('wikisent2.txt')
data = []

for line in txt_file:
    txt = line.strip().lower() # 불러온 문장의 앞뒤 여백 제거 및 소문자 변환 시행 
    data += [re.sub('\'',' ',re.sub(r' [ ]',' ',re.sub(r'[^a-z\' ]','',txt).strip()))]

txt_file.close()
data[:20]
```

```python
dat = ' '.join(data[:500000])
data_split = dat.split(' ')

text_counter = Counter(data_split)

for word, count in text_counter.most_common(1000):
    print(word , '\t', count)
```
